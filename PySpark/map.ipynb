# Aplicar un mapa a un **RDD**

```python
from pyspark import SparkContext
```

```python
sc = SparkContext('local[4]', 'Ejemplo Mapa')
```

```python
numbers = sc.range(1, 6, numSlices=4)
numbers.collect()
```

## El método **map** de un **RDD** aplica una función a cada uno de sus elementos y regresa un nuevo **RDD** (ya que estos son inmutables).

### Recordemos que **Spark** opera mediante **lazy evaluation** y las transformaciones no son evaluadas hasta que ejecutemos una acción (lo que ocurre es que se genera un plan que optimiza la ejecución de las transformaciones cuando se llegue el momento)

```python
squared = numbers.map(lambda x: x ** 2)
```

### A continuación tenemos una **acción** (`collect`) la cuál hace que se ejecute la transformación de la celda anterior

```python
squared.collect()
```

### Este es otro ejemplo de una **acción** (cuenta el número de elementos en el **RDD**)

```python
squared.count()
```
